```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Priya Dharshini R 24BAD092")

filepath = r"C:\Users\DARSHINI\Downloads\archive (1).zip"
data = pd.read_csv(filepath, encoding="latin1")

print(data.head())
print(data.tail())
print(data.info())
print(data.describe())

# Missing values
print(data.isnull().sum())

# Sales calculation
data['Sales'] = data['Quantity'] * data['UnitPrice']

# Top 10 products by sales
top_products = data.groupby('Description')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,8))
sns.lineplot(x=top_products.values, y=top_products.index)
plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Product Name")
plt.show()

plt.figure(figsize=(10,8))
sns.barplot(x=top_products.values, y=top_products.index)
plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Product Name")
plt.show()

print("Success")
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load dataset (ZIP file fix applied)
my_data = pd.read_csv(
    r"C:\Users\DARSHINI\Downloads\archive (2).zip",
    compression='zip',
    encoding="latin1"
)
# Student details
print("Priya Dharshini R 24BAD092")
# Dataset overview
print(my_data.head())
print(my_data.tail())
print(my_data.info())
print(my_data.describe())
print(my_data.isnull())
print(my_data.columns)

# Glucose distribution
plt.figure(figsize=(10,6))
sns.histplot(my_data['Glucose'], bins=30, kde=True)
plt.title("Glucose Level Distribution (Histogram)")
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x=my_data['Glucose'])
plt.title("Glucose Level Distribution (Boxplot)")
plt.xlabel("Glucose Level")
plt.show()

# Age distribution
plt.figure(figsize=(10,6))
sns.histplot(my_data['Age'], bins=30, kde=True)
plt.title("Age Distribution (Histogram)")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x=my_data['Age'])
plt.title("Age Distribution (Boxplot)")
plt.xlabel("Age")
plt.show()

# Correlation between health metrics
plt.figure(figsize=(10,8))
sns.heatmap(my_data.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Health Metrics")
plt.show()
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load dataset
my_data = pd.read_csv(r"C:\Users\DARSHINI\Downloads\archive (3).zip", encoding="latin1")
# Student details
print("Priya Dharshini R 24BAD092")
# Dataset inspection
print(my_data.head())
print(my_data.tail())
print(my_data.info())
print(my_data.describe())
print(my_data.isnull())
print(my_data.columns)
# Scatter plots
plt.figure(figsize=(7,5))
plt.scatter(my_data['area'], my_data['price'], color='blue')
plt.title("Price vs Area")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
plt.figure(figsize=(7,5))
plt.scatter(my_data['bedrooms'], my_data['price'], color='green')
plt.title("Price vs Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.show()
plt.figure(figsize=(7,5))
plt.scatter(my_data['bathrooms'], my_data['price'], color='red')
plt.title("Price vs Bathrooms")
plt.xlabel("Bathrooms")
plt.ylabel("Price")
plt.show()
# Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(
    my_data[['price','area','bedrooms','bathrooms','stories','parking']].corr(),
    annot=True,
    cmap="coolwarm")
plt.title("Correlation Heatmap of Housing Features")
plt.show()
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load dataset
my_data = pd.read_csv(r"C:\Users\DARSHINI\Downloads\archive (4).zip", encoding="latin1", sep="\t")

# Student details
print("Priya Dharshini R 24BAD092")

# Dataset inspection
print(my_data.head())
print(my_data.tail())
print(my_data.info())
print(my_data.describe())
print(my_data.isnull())
print(my_data.columns)
# Create Age column
my_data['Age'] = 2026 - my_data['Year_Birth']

# Bar plots
plt.figure(figsize=(6,4))
sns.barplot(x=my_data['Age'].value_counts().index,
            y=my_data['Age'].value_counts().values)
plt.title("Age Distribution")
plt.show()
plt.figure(figsize=(6,4))
sns.barplot(x=my_data['Income'].value_counts().index[:10],
            y=my_data['Income'].value_counts().values[:10])
plt.title("Income Distribution")
plt.show()
plt.figure(figsize=(6,4))
sns.barplot(x=my_data['MntWines'].value_counts().index[:10],
            y=my_data['MntWines'].value_counts().values[:10])
plt.title("Spending Pattern (Wine)")
plt.show()

# Box plots
plt.figure(figsize=(5,3))
sns.boxplot(x=my_data['Age'])
plt.title("Age Box Plot")
plt.show()
plt.figure(figsize=(5,3))
sns.boxplot(x=my_data['Income'])
plt.title("Income Box Plot")
plt.show()
plt.figure(figsize=(5,3))
sns.boxplot(x=my_data['MntWines'])
plt.title("Spending Box Plot")
plt.show()
```
